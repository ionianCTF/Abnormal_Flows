```python
df=pd.DataFrame()
```

```python
df=pd.read_excel("./Pcap_Labeled.xlsx")
```

```python
df
```

```python
df.drop(['src_ip','dst_ip'], axis=1, inplace=True)
```

```python
print(df.dtypes)
```

```python
df = df.astype(int)
```

```python
X=df.drop(['abnormal_flow'], axis=1)
y=df['abnormal_flow']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = Sequential()
model.add(Dense(12, input_shape=(29,), activation='relu'))
model.add(Dense(24, activation='relu')) 
model.add(Dense(16, activation='relu'))  
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test))
```

```python
test=pd.DataFrame()
test=pd.read_excel("./Test_Pcap.xlsx", )
```

```python
test = test.astype(int)
```

```python
predictions = model.predict(test)
predicted_labels = predictions.round().astype(int)
```

```python
predicted_labels = pd.DataFrame(predicted_labels, columns=['abnormal_flow'])

predicted_labels.to_csv("predicted_abnormal_flow.csv", index=False)
```
